#1: What is the difference between a function and a method in Python?
Ans: In Python, the terms function and method are closely related but have distinct meanings:
 Function
A function is a block of reusable code that performs a specific task and can be called anywhere in your code.

It is defined using the def keyword and can be called independently.

Functions can take parameters, perform operations, and return results.

Example of a function:
def greet(name):
    return f"Hello, {name}!"
    
print(greet("Alice"))
 Method
A method is essentially a function, but it is associated with an object (or an instance of a class). Methods are functions that belong to an object.

Methods are called on objects, and they usually operate on the data contained within that object (often referred to as attributes).

In Python, methods are defined inside a class and always take at least one argument, typically self, which refers to the instance of the object.

Example of a method:
class Person:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        return f"Hello, {self.name}!"
        
person = Person("Alice")
print(person.greet())  # 'greet' is a method
Key Differences:
Scope: Functions are independent, while methods are bound to an object or class.

Syntax: Methods are called on objects using the dot (.) operator (e.g., object.method()), whereas functions are called directly (e.g., function()).

Self: Methods always implicitly receive the object itself as the first parameter (self), which allows them to access and modify the object's state.


#2: Explain the concept of function arguments and parameters in Python
Ans: In Python, function arguments and parameters refer to the values that are passed to and used by a function when it is called. The terms are closely related, but there’s a subtle difference between them. Let’s break down both concepts:

1. Function Parameters
Parameters are the names you define inside the function's parentheses when you define the function.

They act as placeholders for the values (arguments) that will be passed into the function when it is called.

Parameters define what type of data the function expects to receive.
Example of parameters:
def greet(name, age):
    return f"Hello {name}, you are {age} years old.
    In the example above:

name and age are parameters because they are used as placeholders for the actual data that will be passed to the function.

2. Function Arguments
Arguments are the actual values you pass to the function when you call it.

These are the real data that get assigned to the function’s parameters.

Arguments can be literals, variables, or expressions that represent the data to be used inside the function.

Example of arguments:
print(greet("Alice", 30))

In this case:

"Alice" and 30 are arguments because they are the actual values passed to the function when calling greet.
def add_numbers(a, b):   # 'a' and 'b' are parameters
    return a + b

result = add_numbers(5, 10)  # '5' and '10' are arguments
print(result)
Types of Function Arguments in Python
Positional Arguments

These are the most common type of arguments.

They are assigned based on their position (or order) in the function call.

Example:

print(subtract(10, 5))  # 10 is assigned to x, 5 to y

Defining the Function:

greet("Alice")           # Uses the default message: Hello, Alice!

Function with Variable-Length Keyword Arguments (**kwargs)
If you want to allow a function to accept an arbitrary number of keyword arguments, you can use **kwargs. This collects extra named arguments into a dictionary.

#5: . What are iterators in Python and how do they differ from iterables?

Ans: In Python, the terms iterators and iterables are related, but they have distinct meanings and purposes. Understanding the difference is key to effectively working with loops, comprehensions, and other Python constructs.

1. Iterable
An iterable is any Python object that can be looped over (iterated over). It is any object that implements the __iter__() method or has a __getitem__() method that allows access to its elements one by one.

Common examples of iterables:

Lists

Tuples

Dictionaries

Strings

Sets

Files (like open("file.txt"))

Important Points:

An iterable is an object that can be iterated using a loop (e.g., a for loop).

An iterable object must have the __iter__() method, which returns an iterator.

Iterables can be looped through multiple times. When you use a loop on an iterable, a new iterator is created each time.

2. Iterator
An iterator is an object that represents a stream of data. It is an object that not only implements the __iter__() method (making it iterable) but also implements the __next__() method to get the next item from the sequence.

When you call __next__() on an iterator, it returns the next value in the sequence and advances the iterator. When there are no more items left to return, __next__() raises the StopIteration exception to signal that the iteration is complete.

Important Points:

An iterator is an object that produces items one at a time when you call next() on it.

Iterators can be used only once. After all the items are exhausted, they cannot be reused unless you create a new one.

Iterators are lazy—they don’t generate the entire sequence at once, but instead produce items one by one as needed.

#6: Explain the concept of generators in Python and how they are defined.
Ans: Generators are a type of iterable that allow you to generate a sequence of values lazily, meaning values are produced one at a time as they are needed, rather than all at once. This makes generators more memory efficient than using lists or other collections that store all values in memory.

Generators are often used for large datasets, infinite sequences, or when you want to avoid storing entire datasets in memory. They are defined using the yield keyword instead of return.
Generators are defined using a function that contains the yield keyword. When yield is encountered, the function’s state is saved, and the yielded value is returned to the caller. The function's execution is paused at the yield statement, and can be resumed when the generator is called again.

#7: What are the advantages of using generators over regular functions?
Ans: Using generators instead of regular functions (that return full sequences like lists) comes with several important advantages, particularly when it comes to memory efficiency and performance. Here’s a breakdown of the key benefits of using generators:

Memory Efficiency
Generates values one at a time: Regular functions (especially those that return lists) generate all the values at once and store them in memory. This can be problematic when you're dealing with large datasets, as it can consume a lot of memory.

Lazy Evaluation: A generator produces values one at a time only when requested, which means it doesn't store the entire sequence in memory. It yields one value at a time and pauses until the next value is needed, which is much more memory-efficient.
Performance: Generators can be faster than creating and manipulating entire sequences, especially when dealing with large datasets.

Infinite Sequences: Generators can produce infinite sequences without running out of memory.

Simplified Code: Code that requires iterative processes or streams of data is cleaner and more efficient with generators.

Better for Streaming Data: Generators are ideal for real-time data processing where you want to handle one piece of data at a time.

Pausing and Resuming: Generators allow pausing and resuming function execution, which is useful for stateful computation or workflows.

Concurrency and Async Programming: Generators form the basis of asynchronous programming, improving concurrency for IO-bound tasks.

```python
#1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
def sum_of_even_numbers(numbers):
    # Using a list comprehension to filter out even numbers and summing them directly
    even_numbers = [num for num in numbers if num % 2 == 0]
    return sum(even_numbers)

# Example usage
numbers = [47, 11, 42, 13, 56, 8, 23]
result = sum_of_even_numbers(numbers)
print(f"The sum of even numbers is: {result}")

```

```python
#2: . Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(s):
    # Check if input is a string
    if not isinstance(s, str):
        raise ValueError("The input must be a string.")

    # Reverse the string using slicing
    return s[::-1]

# Example usage
input_string = "hello"
reversed_string = reverse_string(input_string)
print(f"The reversed string is: {reversed_string}")
```

```python
#4:  Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    # Handle cases for numbers less than 2
    if number < 2:
        return False
    # Check divisibility from 2 to the square root of the number
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False  # If divisible by any number other than 1 and itself, it's not prime
    return True  # If no divisors were found, it's prime

# Check for prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
```

```python
#5: Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


class FibonacciIterator:
    def __init__(self, terms):
        # Initialize the number of terms to generate
        self.terms = terms
        self.a, self.b = 0, 1  # The first two numbers in the Fibonacci sequence
        self.count = 0  # Counter to keep track of how many terms have been generated

    def __iter__(self):
        # Return the iterator object itself
        return self

    def __next__(self):
        # Stop iteration if we've reached the specified number of terms
        if self.count >= self.terms:
            raise StopIteration

        # Generate the next Fibonacci number and update the terms
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a

# Example usage:
terms = 10  # Number of terms to generate in the Fibonacci sequence
fib_iterator = FibonacciIterator(terms)

# Loop through the iterator and print the Fibonacci numbers
for number in fib_iterator:
    print(number)
```

```python
#6:  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):  # We include the given exponent, so range is from 0 to exponent
        yield 2 ** i  # Yield 2 raised to the power of i

# Example usage
exponent = 5
for power in powers_of_2(exponent):
    print(power)
```

```python
#7: Implement a generator function that reads a file line by line and yields each line as a string
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line  # Yield each line directly
    except FileNotFoundError:
        print(f"The file at {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
file_path = 'example.txt'  # Replace with your actual file path
for line in read_file_line_by_line(file_path):
    print(line, end='')  # Print each line (end='' prevents adding extra newline)
```

```python
#9:  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# List of temperatures in Celsius
celsius_temperatures = [0, 20, 35, 40, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (9/5) * celsius + 32

# Using map() to apply the conversion function to the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the result
print(fahrenheit_temperatures)
```

```python
#10: Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Function to check if a character is a vowel
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

# Input string
input_string = "Hello, World!"

# Use filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(filtered_string)
```
